### Adjusting for lower number of tests. 

When we have low number of test, then the results are easily skeewed. To adjust for this i am using the method described by 3 blue 1 brown (https://www.youtube.com/watch?v=8idr1WZ1A7Q). 
Say you have 10 negatives results out of 10 tests, your negative rate is 100%, howver your real rate is calculated adding a positive and a negative result (12 test, 11 negative, 1 positive), giving you a score of 11/12 or (91%). 



```
#@title Data processing 
df_USA['pos_rate']=df_USA['positive']/df_USA['total']
df_USA['adjusted_pos_rate']=(df_USA['positive']+1)/(df_USA['positive']+2 + df_USA['negative'])

filter_grade= df_USA['grade'].isin(['A', 'B'])
df_USA
```

```
#@title import the data { form-width: "50px" }
url_World='https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
df_World=pd.read_csv(url_World) #importing form a url
 
df_World=df_World.groupby('Country/Region').sum().sort_values(by=['Country/Region'])
df_World.drop(axis=1, labels=['Lat','Long'], inplace=True)
dates= df_World.columns
df_World.reset_index(inplace=True)
df_World = pd.melt(df_World, id_vars='Country/Region',value_vars=dates)
 
df_World = df_World[df_World.value > 200]
df_World['day_since_100_deaths']= df_World.groupby('Country/Region').cumcount()
df_World.rename(columns={'variable':'date', 'value': 'deaths'}, inplace=True)
df_World.head()
 
filter_deaths= 1500 < df_World['deaths'].groupby(df_World['Country/Region']).transform(max)
df_World['new_deaths']= df_World['deaths']-df_World['deaths'].groupby(df_World['Country/Region']).shift(+1).fillna('0').astype('int64')
df_World['moving_new_deaths'] = df_World.groupby('Country/Region')['new_deaths'].transform(lambda x: x.rolling(3, 1).mean())
df_World['new_deaths_over_deaths']= df_World['new_deaths']/df_World['deaths'].groupby(df_World['Country/Region']).shift(+1).fillna('0').astype('int64')
```

```
fig = px.line(df_World,log_y=True, x='day_since_100_deaths', y='deaths', color='Country/Region', hover_name='Country/Region', hover_data=['deaths','new_deaths', 'date'], title='Totale casi per Regione vs Data')
fig.show()
```

```
filter_nations = df_World["Country/Region"].isin(["Italy", "Belgium", "US", "Spain", "United Kingdom", "Germany", "France", "Sweden" ])
fig = px.line(df_World[filter_nations], x='deaths', y='new_deaths_over_deaths', log_x=True,  color='Country/Region', hover_name='CouSwedenegion', hover_data=['deaths', 'new_deaths', 'date'], title='')
fig.show()

# day_since_100_deaths

```
